The original Hamiltonian is 

$$ H ~=~ \tau_1 \tau_2 \tau_3 - \tau_2 \tau_3 \tau_4 +  \tau_1 \tau_2 \tau_3 \tau_4.$$

If you want to quadratise, you need to apply the reduction formula twice, on two different pairs of $\tau$'s. As a first step, the most convenient choice is $\tau_2 \tau_3$. Then you still have a trilinear coupling so you need to collapse another pair of $\tau$'s.  
At the end you will end up with something like:

$$ H' ~=~ \tau_1a_1 - a_1\tau_4 + a_1a_2 + \Lambda  (\tau_2\tau_3 - 2a_1(\tau_2+\tau_3) + 3a_1) + \Lambda  (\tau_1\tau_4 - 2a_2(\tau_1+\tau_4) + 3a_2)  $$

The minimum of the Hamiltonian $H$ can be found just by looking at its structure. The minimum is at $\tau_1 = 0, \tau_2 = \tau_3 = \tau_4=1$ with $H=-1$. Just to make sure, we check explicitly the all the possible combinations of $\tau$'s.



```python
print("t1  t2  t3  t4 -----> Value")
for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                
                
                print(i," ",j," ",k," ",l,"------>",i*j*k - j*k*l + i*j*k*l )
```
